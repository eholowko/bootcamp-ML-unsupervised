### Rozkad SVD

## X - macierz $ m\times n$

rozkładamy macierz na trzy czynniki:


## $ X = U \times \Sigma \times V^t$

, gdzie: $U: m \times m , \Sigma: m \times n, V: n \times n $

- U i V - ortogonalne o wyznacznikach 1 (są to macierze zawierające wektory osobliwe). Są to macierze obrotu.
- $\Sigma$ - macierz wartości osobliwych - $diag(\sigma_1, ..., \sigma_k)$ ,  sigmy uporządkowane malejąco. Macierz jest tego samego wymiary co X zawsze (czyli diagonalna, ale prostokątna). Jest to macierz skalująca.

**Wartości osobliwe** - pierwiastki wartości własnych macierzy $XX^t$

```python
im_r = im[:,:,0]
im_g = im[:,:,1]
im_b = im[:,:,2]
```

### SVD decomposition on R, G, B channels

```python
U_r,s_r,V_r = np.linalg.svd(im_r)
S_r = np.zeros_like(im_r)
np.fill_diagonal(S_r,s_r)
```

```python
U_g,s_g,V_g = np.linalg.svd(im_g)
S_g = np.zeros_like(im_g)
np.fill_diagonal(S_g,s_g)
```

```python
U_b,s_b,V_b = np.linalg.svd(im_b)
S_b = np.zeros_like(im_b)
np.fill_diagonal(S_b,s_b)
```

### Recovering image from k hidden features

```python
k = 150
r_recover = np.dot(np.dot(U_r[:,:k],S_r[:k,:k]),V_r[:k,:])
g_recover = np.dot(np.dot(U_g[:,:k],S_g[:k,:k]),V_g[:k,:])
b_recover = np.dot(np.dot(U_b[:,:k],S_b[:k,:k]),V_b[:k,:])
```

```python
r_recover[r_recover>1] = 1
g_recover[g_recover>1] = 1
b_recover[b_recover>1] = 1
```

```python
r_recover[r_recover < 0] = 0
g_recover[g_recover < 0] = 0
b_recover[b_recover < 0] = 0
```

```python
im_recover = np.stack([r_recover,g_recover,b_recover],axis=2)
```

```python
im_recover.shape
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im_recover)
```

```python
total_size = 0
for m in [U_r[:,:k],U_g[:,:k],U_b[:,:k],s_r[:k],s_g[:k],s_b[:k],V_r[:k,:],V_g[:k,:],V_b[:k,:]]:
    total_size+=m.nbytes
```

## $$U_{new} = X_{new} \cdot V^t \cdot S^{-1}$$

```python
S_inv = np.linalg.inv(S[:k,:k])
```
